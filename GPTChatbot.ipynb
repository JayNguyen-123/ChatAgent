```python
!pip install openai
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))



def get_chat_response(prompt):
  try:
    response = OpenAI.Completion.create(
        engine="text-davinci--003",
        prompt=prompt,
        max_tokens=150, # response length
        temperature=0.7, # controls the random of responses
        n=1 # number of response to generate
    )

    return response.choice[0].text.strip()

  except Exception as e:
    return f"error: {str(e)}"

# Main loop
print("Chatbot: Hello! How can i assist you today?")

while True:
  user_input = input("You: ")
  if user_input.lower() in ["exit", "quit", "bye"]:
    print("Chatbot: Goodbye!")
    break
  chat_prompt = f"User: {user_input}\nChatbot:"
  response = get_chat_response(chat_prompt)
  print("Chatbot: ", response)



```

```python
# Create a chatbot act as an interviewer interviewing candidates for an AI service company

client = OpenAI(
    os.environ.get("OPENAI_API_KEY")

)
def get_completion_from_message(messages, model="gpt-3.5-turbo", temperature=0.7):
  response = client.chat.completions.create(
      model=model,
      messages=messages,
      temperature=temperature
  )
  return response.choice[0].message['content']
```

```python
!pip install jupyter_bokeh

# Create a chatbot act as an interviewer interviewing candidates for an AI service company
# User GUI
import panel as pn
pn.extension()

panels = []

context = [{'role': 'system', 'content':""" I want you to act as an interviewing agent, named Tom,
               for an AI services company.
               You are interviewing candidates, appearing in the interview.
               I want you to only ask questions as the interviewer related to AI.
               Ask one questions at a time."""}]

def collect_messages(_):
  prompt = inp.value_input
  inp.value = ''
  context.append({'role': 'user', 'content': f"{prompt}"})
  response = get_completion_from_message(context)
  context.append({'role': 'assistant', 'content': f"{response}"})
  panels.append(
      pn.Row("User:", pn.pane.Markdown(prompt, width=600))
  )

  panels.append(pn.Row("Assistant: ", pn.pane.Markdown(response, width=600,
                                                       style={'background-color': '#F6F6F6'}))

  )
  return pn.Column(*panels)

# displaying the dashboard
inp = pn.widgets.TextInput(value='Hi', placeholder='Enter text here...')
button_conversation = pn.widgets.Button(name="Chat!")

interactive_conversation = pn.bind(collect_messages, button_conversation)




dashboard= pn.Column(
    inp,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True,
             height=300)
)
dashboard
```
